```python
from skimage import io, data
import matplotlib.pyplot as plt

img=data.chelsea()
rows,cols,dims=img.shape

img2=img.copy()

plt.figure(num='chelsea',figsize=(8,8))
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows):
    for j in range(0,cols):
        img2[i,j,2]=0.5*img2[i,j,2]
        
plt.subplot(1,2,2)
plt.title('modified image')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img=io.imread('C:\\Users/USER/pic/black cat.jpg')
rows,cols,dims=img.shape

img2=np.uint8(np.zeros((rows,cols,3)))
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows):
    for j in range(0,cols):
        sum=0.0
        for k in range(0,3):
            sum=sum+img[i,j,k]
        for k in range(0,3):
            img2[i,j,k]=np.uint8(sum/3)
            
plt.subplot(1,2,2)
plt.title('grey image')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img=io.imread('C:\\Users/USER/pic/black cat.jpg')
rows,cols,dims=img.shape
row2=int(rows/2)
col2=int(cols/2)

img2=img

for i in range(0,row2):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k]=img[i,j,k]
            
for i in range(row2,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k]=img[row2+(-1)*(i-row2),j,k]
            
io.imshow(img2)
io.show()
```
